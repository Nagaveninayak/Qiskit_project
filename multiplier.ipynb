```python
from qiskit import *
```

```python
def adder(inp1, inp2):
    qc = QuantumCircuit(3,2)
    if(inp1 == '1'):
        qc.x(0)
    if(inp2 == '1'):
        qc.x(1)
    
    qc.barrier()
    qc.ccx(0, 1, 2)
    qc.measure(2, 1)
    qc.barrier()
    qc.cx(0, 1)
    qc.measure(1, 0)
    qc.draw()
    
    backend = Aer.get_backend('qasm_simulator')
    job = execute(qc, backend, memory = True)
    output = job.result().get_memory()[0]
    
    return qc, output
```

```python
def multiplier(inp1 ,inp2, inp3, inp4):
    qc = QuantumCircuit(10, 4)
    if(inp1 == '1'):
        qc.x(0)
    if(inp2 == '1'):
        qc.x(1)
    if(inp3 == '1'):
        qc.x(0)
    if(inp4 == '1'):
        qc.x(1)
    
    #A0B0 = P0
    qc.ccx(0, 1, 4)
    qc.measure(4, 0)
    qc.barrier()
    
    #A0B1 + A1B0 = P1
    qc.x(2)
    qc.x(3)
    qc.ccx(0, 3, 5)
    qc.ccx(1, 2, 6)
    qc.cx(5, 6)
    qc.ccx(5, 6, 7)
    qc.measure(6, 1)
    qc.barrier()
    
    #A1B1 + carry = P2 and P3
    qc.ccx(2, 3, 8)
    qc.cx(7, 8)
    qc.ccx(2, 3, 8)
    qc.measure(7, 2)
    qc.measure(8, 3)
    qc.barrier()
    
    #simulators
    backend = Aer.get_backend('qasm_simulator')
    job = execute(qc, backend, memory = True)
    output = job.result().get_memory()[0]
    
    return qc, output
```

```python
for inp1 in ['0', '1']:
    for inp2 in ['0', '1']:
        for inp3 in ['0', '1']:
            for inp4 in ['0', '1']:
                qc, output = multiplier(inp1, inp2, inp3, inp4)
                print('{}{}  * {}{} '.format(inp1,inp2,inp3,inp4), "=", output)
display(qc.draw())
```

```python
inp1 = input('Enter a binary number 1')
inp2 = input('Enter a binary number 2')
qc, output = multiplier(inp1, inp2)
print('{}  * {} '.format(inp1, inp2), "=", output)
qc.draw()
```
